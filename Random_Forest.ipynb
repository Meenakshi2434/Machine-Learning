```python
from sklearn.datasets import load_iris
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
import numpy as np
```

```python
import pandas as pd
```

```python
np.random.seed(0)
```

```python
iris=load_iris()
df=pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

```python
# adding new columns for species
df['species']=pd.Categorical.from_codes(iris.target, iris.target_names)
df.head()
```

```python
# creating test and train data
df['is_Train']=np.random.uniform(0, 1, len(df))<=0.75
df.head()
```

```python
# creating dataframes with test rows and training rows
train, test= df[df['is_Train']==True], df[df['is_Train']==False]
#showing no. of observations for both test and train dataframes
print('Number of Observation in the Training dataframe:', len(train))
print('Number of Observation in the Test dataframe:', len(test))
```

```python
#creating list of features
features= df.columns[:4]
features
```

```python
#converting species to digits
y=pd.factorize(train['species'])[0]
y
```

```python
clf= RandomForestClassifier(n_jobs=2, random_state=0)
clf.fit(train[features], y)
```

```python
#apply trained classifier to test
clf.predict(test[features])
```

```python
# predicting probabilities view
clf.predict_proba(test[features])[0:10]
```

```python
#mapping names for the plants for each predicted plant names
preds= iris.target_names[clf.predict(test[features])]
#view 25
preds[0:25]
```

```python
#viewing the actual species for the first five observations
test['species'].head()
```

```python
#creating confusion matrix
pd.crosstab(test['species'], preds, rownames=['Actual Species'], colnames=['Predicted Species'])
```
